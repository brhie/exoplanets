# Exploring Planets that are potentially habitable
### The data is from NASA Exoplanet Archive(https://exoplanetarchive.ipac.caltech.edu/)<br>The reference for analysis:<br> https://en.wikipedia.org/wiki/Planetary_habitability<br>https://www.liebertpub.com/doi/full/10.1089/ast.2016.1518?journalCode=ast

```python
import pandas as pd
```

```python
df = pd.read_csv('kepler_exoplanets_nasa.csv')
df
```

```python
df.describe()
```

### The filters
- The radius of a potentially habitable exoplanet would range between 0.5 and 1
.5 Earth radi
- The stellar radius; between 0.99 and 1.01 only deviating by 0.01
- The stellar mass; between 0.78 and 1.04

```python
planet_radius = df["koi_prad"]
stellar_radius = df["koi_srad"]
stellar_mass = df["koi_smass"]

confirmed = df[(df["koi_disposition"]=='CONFIRMED') & (planet_radius >= 0.5) & (planet_radius <= 1.5)
               & (stellar_radius >= 0.99) & (stellar_radius <= 1.01) & (stellar_mass >= 0.78) & (stellar_mass <= 1.04)]

print(f'there are {len(confirmed.index)} potentially habitable planets')
confirmed
```

### After filtering, it is listed with the closest from our planet

```python
confirmed.sort_values("koi_dor", ascending=True).reset_index(drop=True)
```

```python
confirmed.plot.scatter(x='koi_dor',
                      y='koi_prad')
```
